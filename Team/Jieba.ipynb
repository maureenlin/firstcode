```python
import jieba
import jieba.analyse  #載入能萃取出文章中關鍵詞的功能
jieba.set_dictionary('jieba.txt.pig')   #載入中文繁體字詞庫

##########################################################################
#若要使用自建辭典（ex：台語、日文...），則要新增一個純文字檔建立自定義詞庫，格式如下：
#  userdict.txt

#  行袂開跤 2 v
#  袂當 4 d
#  袂記 4 v
#  袂有 4 d
#  唱著 4 v
# （自定義詞  權重  詞性（不用寫也沒關係））

#建完後，在載入jieba字典後的下一行輸入： jieba.load_userdict("userdict.txt")
#即可匯入自定義辭典
##########################################################################
count = {}
content = open('test.txt', 'rb').read()   #載入要被斷詞的文件
words = jieba.cut(content, cut_all=False) #cut(斷詞內容,cut_all=True的話會幾乎一個字一個字斷)
for word in words:
    if word in count:
      count[word] += 1
    else:
      count[word] = 1
    print(word)
    

#tags = jieba.analyse.extract_tags(content, 5)  #找出被斷詞文件中5個分數最高的個詞 ????????????????????
#for tag in tags:
    #print(tag)

for i in count:
    print("%s,%d\n"%(i,count[i]))
```
