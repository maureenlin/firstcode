```python
import csv  #載入專門開csv檔案的套件
import jieba
import jieba.analyse  #載入能萃取出文章中關鍵詞的功能
jieba.set_dictionary('dict.txt')

pen = []                                     #分類詞庫
pens = open('pen.txt','r').readlines()
for p in pens:
    pen.append(p.strip())
    
card = []                                    #分類詞庫
cards = open('card.txt','r').readlines()
for c in cards:
    card.append(c.strip())  

floder = []                                  #分類詞庫
floders = open('floder.txt','r').readlines()
for f in floders:
    floder.append(f.strip())

paper = []                                   #分類詞庫
papers = open('paper.txt','r').readlines()
for p in papers:
    paper.append(p.strip())

stick = []                                   #分類詞庫
sticks = open('stick.txt','r').readlines()
for s in sticks:
    stick.append(s.strip())

stop = []                                    #停用詞詞庫
stopkey = open('stop.txt','r').readlines()
for s in stopkey:
    stop.append(s.strip())
```

```python
content = []   #裝最後分詞完的結果
contents = open('POS5007_DEMO.csv', 'r').readlines()    #開啟檔案
for row in csv.DictReader(contents):    #csv.DictReader函式可依欄位名開啟特定欄位
    words = jieba.cut(row['PLU_Name'].replace(' ',''), cut_all=False)    #只讓'PLU_Name'此欄位下的內容做斷詞
    temp = []    #每個詞斷出的結果
    category = []
    for word in words:
        if word not in stop:
            temp.append(word)              
        
        if word in card :          
            category.append('卡片')
        if word in pen :          
            category.append('筆')
        if word in floder :       
            category.append('資料夾')
        if word in paper :         
            category.append('紙')
        if word in stick :         
            category.append('貼紙')
          
    if category:        #如果category不為空字串的話 （空字串在python if條件式下為false）
        content.append(temp+category)
        
    else:
        content.append(temp)
   
    
content





```
