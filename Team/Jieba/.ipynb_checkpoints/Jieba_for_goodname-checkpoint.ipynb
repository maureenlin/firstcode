```python
import csv  #載入專門開csv檔案的套件
import jieba
import jieba.analyse  #載入能萃取出文章中關鍵詞的功能
#jieba.set_dictionary('jieba.txt')   #載入中文繁體字詞庫
jieba.set_dictionary('dict.txt')

stop = []
stopkey = open('stop.txt','r').readlines()
for s in stopkey:
    stop.append(s.strip())
```

```python
content = []   #裝最後分詞完的結果
contents = open('POS5007_DEMO.csv', 'r').readlines()    #開啟檔案
for row in csv.DictReader(contents):    #csv.DictReader函式可依欄位名開啟特定欄位
    words = jieba.cut(row['PLU_Name'], cut_all=False)    #只讓'PLU_Name'此欄位下的內容做斷詞
    temp = []    #每個詞斷出的結果
    for word in words:
        if word not in stop:
            temp.append(word)
    content.append(temp)
content
```
