```python
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text = open('104R.csv').read()  #讀檔
```

```python
wordcloud = WordCloud(max_font_size=500).generate(text)   
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
